### Three step sizes, same valley
Run the same descent with three learning rates and watch the final loss.

```python
def L(w): return (w-3)**2
def descent(w0, lr, n):
    w = w0
    for _ in range(n): w = w - lr*2*(w-3)
    return w
for lr in (0.02, 0.15, 0.62):
    w = descent(-1.0, lr, 8)
    tag = 'crawls' if L(w) > 4 else ('DIVERGES' if L(w) > 100 else 'converges')
    print(f'lr={lr:<5}  final loss={L(w):10.2f}   -> {tag}')
check('lr=0.15 converged (final loss small)', lambda: L(descent(-1.0, 0.15, 40)) < 0.01)
check('lr=0.62 has NOT converged in 8 steps', lambda: L(descent(-1.0, 0.62, 8)) > 0.01)
```
