### The loss and its slope
Our toy loss is `L(w) = (w − 3)²` — a valley with its bottom at w = 3.

```python
def L(w):      return (w - 3)**2      # how wrong the model is at weight w
def slope(w):  return 2*(w - 3)       # the derivative: the slope under your feet
print('L(-1) =', L(-1), '  slope(-1) =', slope(-1))
check('at w=-1 the ground slopes down to the right (negative slope)', lambda: slope(-1) < 0)
```

### Roll downhill
One step is `w ← w − lr · slope(w)`. Watch the loss fall.

```python
w, lr = -1.0, 0.15
for step in range(6):
    print(f'step {step}:  w={w:+.3f}  loss={L(w):.3f}  slope={slope(w):+.3f}')
    w = w - lr*slope(w)
check('after 6 steps the ball is near the valley (w ~ 3)', lambda: abs(w-3) < 1.5)
```
