## Part A — Learn
Build the learner and watch its error fall as you feed it more labelled points.

```python
# A hidden 'true' survival rectangle (temp 18-28, humidity 45-72). The learner never sees it.
TRUE = (18, 28, 45, 72)   # (tmin, tmax, hmin, hmax)
def truth(p): return TRUE[0]<=p[0]<=TRUE[1] and TRUE[2]<=p[1]<=TRUE[3]

def sample(n, seed=0):
    r=random.Random(seed)
    return [(r.uniform(10,40), r.uniform(20,90)) for _ in range(n)]

def learn_and_error(m, seed=0, test_n=4000):
    data=sample(m, seed)
    pos=[p for p in data if truth(p)]
    if not pos: return 1.0
    xs=[p[0] for p in pos]; ys=[p[1] for p in pos]
    box=(min(xs),max(xs),min(ys),max(ys))          # tightest rectangle on the ✓ points
    inbox=lambda p: box[0]<=p[0]<=box[1] and box[2]<=p[1]<=box[3]
    test=sample(test_n, seed+999)
    err=sum(1 for p in test if inbox(p)!=truth(p))/test_n
    return err

for m in (10, 40, 160, 640):
    print(f'm={m:>4}   generalization error ~ {learn_and_error(m):.3f}')
check('error falls as m grows (640 samples beats 10)', lambda: learn_and_error(640) < learn_and_error(10))
```
